```python
import itertools

# Generate all possible (Product, Variety/Size, Brand) combinations
all_combinations = []
for product in categories:
    for variety in size_dict[product]:
        for brand in brands_dict[product]:
            all_combinations.append((product, variety, brand))
```

```python
## Checking if the combinations have any duplicates
if len(all_combinations) != len(set(all_combinations)):
    raise ValueError("Duplicate combinations found in all_combinations")
```

```python
print(f"Total combinations: {len(all_combinations)}")
```

```python
# If total combinations < n, sample with replacement; else, sample without replacement

if len(all_combinations) >= n:
    sampled_combinations = random.sample(all_combinations, n)
else:
    sampled_combinations = random.choices(all_combinations, k=n)

for combo in sampled_combinations:
    product, variety, brand = combo
    # Get price range and sample a price
    price_min, price_max = price_ranges[product][variety]
    price = random.randint(price_min, price_max)
    # Get base carbon footprints
    prod_cf = production_carbon_footprint[product][variety]
    pack_cf = packaging_carbon_footprint[product][variety]
    log_cf = logistics_carbon_footprint[product][variety]
    # Get brand efficiency factor (default to 1.0 if not found)
    brand_factor = brand_efficiency_factors.get(brand, 1.0)
    # Adjust carbon footprints by brand factor
    prod_cf_adj = round(prod_cf * brand_factor, 2)
    pack_cf_adj = round(pack_cf * brand_factor, 2)
    log_cf_adj = round(log_cf * brand_factor, 2)
    total_cf = round(prod_cf_adj + pack_cf_adj + log_cf_adj, 2)
    # Store the record
    products.append({
        "itemID": item_id,
        "Product": product,
        "Variety/Size": variety,
        "Brand": brand,
        "Price": price,
        "Production(kg CO2e)": prod_cf_adj,
        "Packaging(kg CO2e)": pack_cf_adj,
        "Logistics(kg CO2e)": log_cf_adj,
        "Total(kg CO2e)": total_cf
    })
    item_id += 1

# Create DataFrame
synthetic_df = pd.DataFrame(products)
synthetic_df.head()
```
